```python
# with Exceptional Error Handling
try:
  with open("sample.txt","r") as file:
    content =file.read()
    print(content)
except FileNotFoundError:
  print("File Not Found")
```

```python
def load_recipes(file_path):
    recipe_dict = {}

    try:
        with open(file_path, "r") as file:
            lines = [line.strip() for line in file if line.strip() != ""]

        i = 0
        while i < len(lines):
            name = lines[i]
            ingrediants = lines[i+1].replace("Ingrediants:", "").strip()
            instructions = lines[i+2].replace("Instructions:", "").strip()

            recipe_dict[name] = {
                "ingrediants": ingrediants,
                "instruction": instructions
            }

            i += 3

        return recipe_dict

    except FileNotFoundError:
        print("File not found")
        return {}


def show_menu():
    print("\n-------- Recipe Viewer Menu------")
    print("1. View Recipe by Name")
    print("2. List All Recipes")
    print("3. Exit")


def view_recipe(recipes):
    name = input("Enter the name of the recipe: ")

    if name in recipes:
        print(f"\n-------- Recipe {name} Details------")
        print(f"Name: {name}")
        print(f"Ingrediants: {recipes[name]['ingrediants']}")
        print(f"Instructions: {recipes[name]['instruction']}")
    else:
        print("Recipe not found")


recipe_file = "Recipes.txt"
recipes = load_recipes(recipe_file)

while True:
    show_menu()
    choice = input("Enter your choice (1/2/3): ")

    if choice == "1":
        view_recipe(recipes)

    elif choice == "2":
        print("\n-------- All Recipes------")
        for name in recipes:
            print(name)

    elif choice == "3":
        print("Exiting Recipe Viewer. Goodbye!")
        break

    else:
        print("Invalid choice. Please try again.")
```
