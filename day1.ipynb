```python
while True:
   x_input=input("Please Enter first Number or type exit to quite: ")

   if x_input.lower() =='exit':
    print("Thanks for using calculator")
    break
   y_input=input("Please Enter Second Number or type exit if you want: ")
   if y_input.lower() =='exit':
    print("Thanks for using calculator")
    break

   x=float(x_input)
   y=float(y_input)


   print("\n ---Select Operation you want to perform :")
   print("1. Addition (+)")
   print("2. Subtraction (-)")
   print("3. Multiplication (*)")
   print("4. Division (/)")
   print("5. Exponentional (**)")


   choice= input("Enter your choice Number: ")


   result =None
   if choice == '1':
     result = x+y
   elif choice =='2':
     result =x-y
   elif choice =='3':
     result = x*y
   elif choice == '4':
     if y!=0:

      result =x/y
     else:
       print("Divison with zero is not allowed")
       continue

   elif choice =='5':
      result =x**y
   else:
     print("You enter number is out of choice sorry!")
     continue
   if result is not None:

     print("Result : ",result)
```

**Build a simple python game  Number Guessing**

```python
import random

secret_number = random.randint(1,10)
attempts=3

while attempts>0:
  guess =int(input("Take a Guess"))
  if guess == secret_number:
    print("COngragulations You guess correct number which is :", secret_number)
    break
  elif guess<secret_number:
    print("Try high number:")
  else:
    print("Try low number")

  attempts -=1
if attempts ==0:
  print("Sorry you can not guess nmumber! whihc is ",secret_number)
```
