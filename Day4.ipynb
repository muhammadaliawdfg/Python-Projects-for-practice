what are sets?

```python
# A set is an unordered collection of unique items. Sets are mutable, but they do not allow duplicate values.
#  syntax : set_name = {item1, item2, item3}  for emp[ty] empty_set = set()
my_set = {1,2,3,3}
print(my_set)

ingrediants = {"salt","pepper","oil"}
print(ingrediants)
ingrediants.add("eggs")
print(ingrediants)
ingrediants.remove("salt")
print(ingrediants)
```

```python
# set operations union, intersaction , difference
# union :Combines elements of both sets (no duplicates).
# Intersaction:Returns common elements in both sets.
# DIfference:Returns elements in first set but not in second set.
set_a ={"sugar","butter","milk"}
set_b ={"sugar","eggs"}
# print(set_a.union(set_b))
# print(set_a.intersaction(set_b))
# print(set_a.difference(set_b))
print(set_a | set_b)
print(set_a & set_b)
print(set_a - set_b)
```

```python
# An exception in Python is an error that occurs during the execution of a program, which interrupts the normal flow of the program
# try:
#     # code that may raise an exception
# except ExceptionType:
#     # code to handle the exception
# else:
#  Execute if no exception occurs
# finally
#  Always execute, even if an exception occurs

try:
  num =int(input("Enter a number: "))
  result =10/num
  print("Result: ", result)
except ZeroDivisionError:
    print("Error: Division by zero")
else:
  print("no Exception occur the result is: ", result)
finally:
  print("The block excecuted. Program Ended")
```

```python
# we also use a multiple exceptional handling combioned
try:
  num =int(input("Enter a number: "))
  result =10/num
  print("Result: ", result)
except (ZeroDivisionError, ValueError):
    print("Error: Division by zero or invalid Input")
```

```python
# Safe Calculator
def add(x,y):
  return x+y
def subtract(x,y):
  return x-y
def multiplication(x,y):
  if x<0 or y<0:
    raise ValueError("Negative values are not allowed")
  elif x==0 or y==0:
    raise ValueError("Zero values are not allowed")
  return x*y
def division(x,y):
  if y==0:
    raise ZeroDivisionError("Division by zero is not allowed")
  return x/y
def show_menu():
  print("\n ---- Safe Calculator ----")
  print("1. Addition")
  print("2. Subtract")
  print("3. Multiply")
  print("4. Divide")


while True:
  show_menu()
  choice = input("Enter your choice (1/2/3/4): ")
  if choice == "5":
    print("Exiting the Calculator. Goodbye")
    break
  try:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    if choice == "1":
      print("Result: ", add(num1,num2))
    elif choice == "2":
      print("Result: ", subtract(num1,num2))
    elif choice == "3":
      print("Result: ", multiplication(num1,num2))
    elif choice == "4":
      print("Result: ", division(num1,num2))
    else:
      print("Invalid Choice. Please Select a valid option.")
  except ValueError as e:
    print("Invalid Input. Please enter valid numbers.",e)
  except ZeroDivisionError:
    print("Error: Division by zero is not allowed")


```
