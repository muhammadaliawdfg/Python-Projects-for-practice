```python
import requests

def get_detailed_weather(city_name):
    # Step 1: Coordinates nikalna
    geo_url = f"https://geocoding-api.open-meteo.com/v1/search?name={city_name}&count=1&format=json"

    try:
        geo_res = requests.get(geo_url).json()

        if 'results' in geo_res:
            lat = geo_res['results'][0]['latitude']
            lon = geo_res['results'][0]['longitude']

            # Step 2: Weather + Rain Chance fetch karna
            weather_url = f"https://api.open-meteo.com/v1/forecast?latitude={lat}&longitude={lon}&current_weather=true&hourly=precipitation_probability&forecast_days=1"
            res = requests.get(weather_url).json()

            current = res['current_weather']
            # Abhi ke waqt ka rain probability
            rain_chance = res['hourly']['precipitation_probability'][0]

            print(f"\n--- {city_name.capitalize()} Detailed Weather ---")
            print(f"Temperature : {current['temperature']}°C")
            print(f"Windspeed   : {current['windspeed']} km/h")
            print(f"Rain Chance : {rain_chance}%")

            if rain_chance > 50:
                print("Take Umbrella, the chance of rain is heavy!")
            else:
                print("weather is clean, very less chance of Rain.")
        else:
            print(f"'{city_name}' this is not found.")

    except Exception as e:
        print("check your internet the have issue.")

# --- Runtime User Input ---
user_city = input("Enter the name of city you want to see weather ")

if user_city.strip(): # Check ke user ne khali enter na dabaya ho
    get_detailed_weather(user_city)
else:
    print("you enter nothing!")
```
