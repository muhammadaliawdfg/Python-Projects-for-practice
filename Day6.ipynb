```python
# List Comprehension provide a concise way to create and manipulate list using a signle line of code more readable and faster than traditional loops
# List Comprehension Python ka ek short aur smart tareeqa hai list banane ka ek hi line me.
# synatx
# [expression Har item par kya operation karna hai(x**2)  for item in iterable =(Loop kis cheez par chalana hai) if condition (optional fo filter)]
# order
# expression → for → if
squares =[x**2 for x in range(10)]
print(squares)
```

```python
# odd squares
odd_squares = [x**2 for x in range(10) if x%2 ==0]
print(odd_squares)
```

```python
# Filtering item with list Comprehensions
names = ["ali","ahmed","jawad","awan"]
short_name= [name for name in names if len(name)<=5]
print(short_name)
```

```python
student_scores =input("Enter student scores seprated by commas: ")
scores = [int(score) for score in student_scores.split(",")]
# Assign Grades using List Comprehension
grades = [
    "A" if score >= 90 else
    "B" if score >= 80 else
    "C" if score >= 70 else
    "D" if score >= 60 else
    "F"
    for score in scores
]
print(grades)

# Filtering Passing and failing students
passing_students = [score for score in scores if score>=60]
failing_students = [score for score in scores if score<60]
# Print Results
print("Passing Students:", passing_students)
print("Failing Students:", failing_students)
# step 4 print Results
print("\n----- Student Grades ------")
for i, (score, grade) in enumerate(zip(scores, grades), start =1):
    print(f"Student {i}: Score = {score}, Grade = {grade}")
print("\n------- Passing and Failing Students-----")
print("Passsing Students :" , passing_students)
print("Failing Students: " , failing_students)
```

```python
import random, string

def generate_password(length=12):
    if length < 4:
        raise ValueError("Password length should be at least 4 characters.")

    uppercase_letters = string.ascii_uppercase
    lowercase_letters = string.ascii_lowercase
    digits = string.digits
    special_characters = "!@#$%^&*(_+-{}[]|<>?)"

    password = [
        random.choice(uppercase_letters),
        random.choice(lowercase_letters),
        random.choice(digits),
        random.choice(special_characters)
    ]

    all_characters = uppercase_letters + lowercase_letters + digits + special_characters
    password += random.choices(all_characters, k=length-4)

    random.shuffle(password)
    return "".join(password)

while True:
    try:
        length = int(input("Enter password length (minimum 4): "))
        password = generate_password(length)
        print("Generated Password:", password)

        again = input("Do you want another password? (y/n): ").lower()
        if again != 'y':
            print("Program Ended.")
            break

    except ValueError as e:
        print("Error:", e)
```
